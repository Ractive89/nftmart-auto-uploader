```python
import json

load ={}
with open("data.json",'r') as load_f:
    load = json.load(load_f)

print(load[0]['name'])
print(load[0]['filePath'])
print(load[0]['description'])
print(load[0]['properties'][0]['color'])
print(load[0]['properties'][1]['type'])
print(load[0]['price'])
out={"collection":"Void Explode",'lists':[]}
for i in load[:5]:
    out['lists'].append({"name": i['name'],
                        "filePath": i['filePath'],
                        "description": i['description'],
                        "properties": { "color": i['properties'][0]['color'] , "type": i['properties'][1]['type'] },
                        "price": i['price']
                       })
print(json.dumps(out, indent=4, sort_keys=True))
with open('data1.json', 'w') as outfile:
    outfile.write(out)
```
